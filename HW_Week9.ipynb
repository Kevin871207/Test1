```python
import numpy as np
import pandas as pd

train = pd.read_csv('forestfires.csv')
test = pd.read_csv('test.csv')

print('相關係數為:')
print(train.corr()[['area']])
```

## 所以可以看出temp跟DMC的相關係數最接近1或-1因此他們兩個較相關,因此選擇他們兩個

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

train_data = train[['temp']]
train_label = train[['area']]
test_data = test[['temp']]
test_label = test[['area']]

lm = LinearRegression()
lm.fit(train_data, train_label)
print("迴歸係數:{}  截距:{}".format(lm.coef_,lm.intercept_))
predict_label = lm.predict(test_data)

plt.scatter(train_data, train_label, color='blue', label='train_data_train_label')
plt.scatter(test_data, test_label, color='green',label='test_data_test_label')
plt.plot(test_data, predict_label, color='red', marker="o", markersize=10,label='test_data_predict_label')
plt.xlabel('temp')
plt.ylabel('area')
plt.title('temp-area')
plt.legend()
plt.grid()
plt.show()
#######################
print("---------------------------------")
train_data = train[['DMC']]
train_label = train[['area']]
test_data = test[['DMC']]
test_label = test[['area']]

lm = LinearRegression()
lm.fit(train_data, train_label)
print("迴歸係數:{}  截距:{}".format(lm.coef_,lm.intercept_))
predict_label = lm.predict(test_data)

plt.scatter(train_data, train_label, color='blue', label='train_data_train_label')
plt.scatter(test_data, test_label, color='green',label='test_data_test_label')
plt.plot(test_data, predict_label, color='red', marker="o", markersize=10,label='test_data_predict_label')
plt.xlabel('DMC')
plt.ylabel('area')
plt.title('DMC-area')
plt.legend()
plt.grid()
plt.show()
```
